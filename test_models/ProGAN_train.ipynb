```python
from progan import ProGAN, setup_data, IN_CHANNEL
import torch as th
import os

# select the device to be used for training
# device = th.device("cuda" if th.cuda.is_available() else "cpu")
device = th.device('cuda:0') if th.cuda.is_available() else th.device('cpu')

name = 'Ursus_arctos_Linnaeus'
output_path = f'../output/{name}_progan'

if not os.path.exists(f'{output_path}'):
    os.makedirs(f'{output_path}')

log_dir =    f'{output_path}/logs/'
sample_dir = f'{output_path}/samples/'
save_dir =   f'{output_path}/models/'

data_path = f"../data/RGBA256"
# some parameters:
depth = 7
START_DEPTH = 2
START_EPOCH = 15
# hyper-parameters per depth (resolution)
# 4x4, 8x8, 16x16, 32x32, 64x64, 128x128, 256x256, 512x512
num_epochs = [70, 70, 70, 70, 70, 100, 100]
fade_ins = [10, 20, 30, 40, 50, 100, 100]
batch_sizes = depth*[8]
latent_size = 256
# get the data. Ignore the test data and their classes
dataset= setup_data(data_path)
print(f'data num:{len(dataset)}')
print(dataset[0][0].shape)
IN_CHANNEL= dataset[0][0].shape[0]
print(f"channel: {IN_CHANNEL}")
# dataset1 = np.load(path + '/Data/Train_5cm_norm.npz')
# dataset2 = dataset1['arr_0']
# print(dataset2.shape)
# dataset = TensorDataset(*dataset2)
# ======================================================================
# This line creates the PRO-GAN
# ======================================================================
# pro_gan = ConditionalProGAN(num_classes=10, depth=depth, 
#                                latent_size=latent_size, device=device)
pro_gan = ProGAN(depth=depth, latent_size=latent_size, device=device)
# ======================================================================
# ======================================================================
# This line trains the PRO-GAN
# ======================================================================
pro_gan.train(
    dataset=dataset,
    epochs=num_epochs,
    fade_in_percentage=fade_ins,
    batch_sizes=batch_sizes,
    start_epoch= START_EPOCH,
    start_depth= START_DEPTH,
    save_dir=save_dir,
    log_dir=log_dir,
    sample_dir=sample_dir
)
# ======================================================================
```
